```python
import pexpect
import time
import sys
import re
```

```python
#ENM3 -> 10.23.202.28 PW : "Enm#2021"
#ENM5 -> 10.167.30.28 PW : "Enm#2021"

child = pexpect.spawn('ssh -tt -oProxyJump="root@150.236.203.198:22,root@127.0.0.1:8008" eidtriyul@10.167.30.28', timeout=30, encoding='utf-8', logfile=sys.stdout)

child.expect("root@150.236.203.198's password: ")
child.sendline('PMT@2022')
child.expect("root@127.0.0.1's password: ")
child.sendline('PMT@2022')
# time.sleep(5)

child.expect("eidtriyul@10.167.30.28's password: ")
child.sendline('Enm#2021')

# time.sleep(5)
child.expect("eidtriyul: You must SSH using username and password to schedule CRON jobs.")

child.sendline('cd /ericsson/pmic1/CELLTRACE/')
child.expect("")

child.sendline('ls -d * */*')

try :
    child.expect(pexpect.EOF)

except:
    child.close()
    test1 = str(child.before)
```
