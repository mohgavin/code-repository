```python
cellrebel = pd.read_csv('rsrp_cell_rebel.csv')
kabupaten = gpd.read_file('Polygon Kecamatan_Fajar/Kab_Kec_Combined_June_2021V1.TAB')
```

```python
pd_to_gpd = gpd.GeoDataFrame(cellrebel, geometry=gpd.points_from_xy(cellrebel.longitude, cellrebel.latitude), crs=4326)
```

```python
pd_to_gpd.crs
```

```python
kabupaten.crs
```

```python
# pd_to_gpd = pd_to_gpd.within(karawang)
# pd_to_gpd['test'] = pd_to_gpd.sjoin(karawang, how="inner", predicate='within')
points_within = pd_to_gpd.sjoin(kabupaten, how='inner', predicate='within')
```

```python
# feather = pd.DataFrame(points_within.drop(columns='geometry').reset_index())
# feather.to_csv('cellrebel_mapping_30032023.csv', index=False)
```

```python
points_within = points_within[~points_within['reference_signal_received_power'].isnull()]
```

```python
points_within
```

```python
pivot = pd.pivot_table(points_within , index= ['Prov_Kab_Kec', 'kabupaten'],
                        columns=['mobile_operator'], values=['reference_signal_received_power'],aggfunc = np.mean)
```

```python
pivot.reset_index(inplace=True, col_level=1, allow_duplicates=False)
```

```python
pivot = pivot.droplevel(level=0, axis=1)
```

```python
pivot
```

```python
test = pivot.merge(kabupaten[['Prov_Kab_Kec', 'geometry']], how='left', on='Prov_Kab_Kec')
```

```python
test1 = gpd.GeoDataFrame(test, geometry='geometry', crs=4326)
```
