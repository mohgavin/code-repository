```python
import os
import zipfile

for files in os.listdir('/home/nivag/2023 - Linux/Compile-ZIP'):
    print(files)
    fullpath = os.path.join('/home/nivag/2023 - Linux/Compile-ZIP', files)

    # Extract the zip file
    with zipfile.ZipFile(fullpath, 'r') as zip_ref:
        zip_ref.extractall('/home/nivag/2023 - Linux/Compile-MDT')

    # Get the name of the zip file without the extension
    zip_name = os.path.splitext(files)[0]
    zip_name = zip_name.replace('_mdt_result', '')

    # Rename the extracted file
    extracted_file = os.path.join('/home/nivag/2023 - Linux/Compile-MDT', "mdt_result.csv")
    new_file_name = os.path.join('/home/nivag/2023 - Linux/Compile-MDT', f"mdt_result_{zip_name}.csv")
    os.rename(extracted_file, new_file_name)
```
