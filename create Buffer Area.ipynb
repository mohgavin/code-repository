```python
import pandas as pd
import geopandas as gpd
```

```python
inbuilding = pd.read_excel('sample-folder/Renewal Pico_EID_10082023.xlsx')
inbuilding = inbuilding.astype(str)
inbuilding = gpd.GeoDataFrame(inbuilding, geometry=gpd.points_from_xy(inbuilding['Long'], inbuilding['Lat']), crs=4326)
```

```python
inbuilding['geometry'] = inbuilding['geometry'].buffer(0.0008985/3)
```

```python
inbuilding['WKT'] = inbuilding['geometry']
inbuilding.rename(columns={'Sitename':'polygon_name'}, inplace=True)
```

```python
{c: i for i, c in enumerate(inbuilding.columns)}
```

```python
inbuilding = inbuilding[['polygon_name', 'WKT']]
```
